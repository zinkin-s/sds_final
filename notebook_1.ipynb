```python
import pandas as pd
import numpy as np

df_okved_2007 = pd.read_csv('data/sheet_1.csv')
df_okved_2007.columns
```

```python
df_okved_2 = pd.read_csv('data/sheet_2.csv')
df_okved_2.head()
```

```python
sal_1 = df_okved_2007[['Год', 'Всего по  экономике']]
sal_2 = df_okved_2[['Год', 'Всего']]
sal_2 = sal_2.rename(columns={'Всего': 'Всего по  экономике'})
table = pd.concat([sal_1, sal_2], ignore_index=True)
table
```

```python
inflation = pd.read_csv('data/inflation.csv')
```

```python
inflation = inflation.set_index('Год').sort_index()
z = 100 + inflation['Всего']
table['ИПЦ'] = z
table
```

```python
xs = list(table['Всего по  экономике'])
res = []
res.append(2223 / 1523 * 100)
for i in range(1, len(xs)):
    res.append(xs[i]/xs[i-1]*100)

res = np.array(res)
table['ИНЗП к пред.году'] = res
table['ИРЗП к пред. году'] = table['ИНЗП к пред.году'] / table['ИПЦ'] * 100

arr = []
ys = list(table['ИРЗП к пред. году'])
t = 1
for j in range(len(ys)):
    t = t * (ys[j]/100)
    arr.append(t)
real_sal_temp = np.array(arr) * 100
table['ИРЗП к базовому году'] = real_sal_temp
cip = list(table['ИПЦ'])
t = 1
cip_rate = []
for i in range(len(cip)):
    t = t*cip[i]/100
    cip_rate.append(t)
cip_rate = np.array(cip_rate) * 100
table['ИПЦ к базовому году'] = cip_rate
table['РЗП к базовому году'] = table['Всего по  экономике'] / (cip_rate / 100)

table
```

```python
def salary_index(df, ind_selector):
    res = []
    arr = list(df[ind_selector])
    for i in range(1, len(arr)):
        res.append(arr[i]/arr[i-1]*100)
    data = pd.DataFrame()
    data['НСЗП'] = df[ind_selector].iloc[1:]
   
    res = np.array(res)
    data['ИНЗП к пред.году'] = res
    z = table['ИПЦ'][1:17]
    data = data.merge(z, left_index=True, right_index=True)
    data['ИРЗП к пред.году'] = data['ИНЗП к пред.году'] / data['ИПЦ'] * 100

    arr = []
    ys = list(data['ИРЗП к пред.году'])
    t = 1
    for j in range(len(ys)):
        t = t * (ys[j] / 100)
        arr.append(t)
    real_sal_temp = np.array(arr) * 100
    data['ИРЗП к базовому году'] = real_sal_temp
    return data


selector = 'сельское хозяйство, охота и лесное хозяйство'
df = salary_index(industry_okved, selector)
df[['ИНЗП к пред.году', 'ИРЗП к пред.году']].plot(kind='line')
```
